## Task 10

```python
revenue_wise=df.sort_values(by='revenue', ascending=False)
fist_fivehundred=revenue_wise.head(500)
fist_fivehundred
```

```python
print(fist_fivehundred['budget'].describe())
print(fist_fivehundred['budget'].describe().apply(lambda p: '%.2f' % p))
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(fist_fivehundred.budget, color="g");
```

```python
q1=fist_fivehundred['budget'].describe()['25%']
q3=fist_fivehundred['budget'].describe()['75%']
iqr=q3-q1

upper_bound=q3+1.5*iqr
fist_fivehundred['budget']=np.where(fist_fivehundred['budget']>upper_bound, upper_bound, 
                                       fist_fivehundred['budget'])
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(fist_fivehundred.budget, color="g");
```

```python
print(fist_fivehundred['revenue'].describe())
print(fist_fivehundred['revenue'].describe().apply(lambda p: '%.2f' % p))
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(fist_fivehundred.revenue, color="g");
```

```python
q1=fist_fivehundred['revenue'].describe()['25%']
q3=fist_fivehundred['revenue'].describe()['75%']
iqr=q3-q1

upper_bound=q3+1.5*iqr
fist_fivehundred['revenue']=np.where(fist_fivehundred['revenue']>upper_bound, upper_bound, 
                                       fist_fivehundred['revenue'])
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(fist_fivehundred.revenue);
```

```python
print(fist_fivehundred['runtime'].describe())
print(fist_fivehundred['runtime'].describe().apply(lambda p: '%.2f' % p))
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(fist_fivehundred.runtime, color="y");
```

```python
q1=fist_fivehundred['runtime'].describe()['25%']
q3=fist_fivehundred['runtime'].describe()['75%']

iqr=q3-q1
upper_bound=q3+1.5*iqr

fist_fivehundred['runtime']=np.where(fist_fivehundred['runtime']>upper_bound, upper_bound,
                                     fist_fivehundred['runtime'])
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(fist_fivehundred.runtime, color="y");
```
